```python
import numpy as np
import pandas as pd
import scipy.stats as st
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as mp
infile = 'Source_Data/merged_counties_2_year_merged.csv'
```

```python
data = pd.read_csv(infile)
data.tail()
```

```python
data.count()
```

```python
data.plot(x = 'obesity', y = 'Age Adjusted Rate', kind='scatter', alpha = 0.2, color='darkcyan' )
```

```python
data.columns
```

```python
rmat = data.corr(method='pearson')
```

```python
allobs = data.dropna(subset=['Age Adjusted Rate'])
len(allobs)
```

```python
nmat = pd.DataFrame(allobs.count())
nmat = nmat.rename(columns = {0:'count'})
```

```python
nmat.head()
```

```python
nmat = nmat.drop(['county','county_and_state_x','county_and_state_y','state'], axis = 0)
nmat
```

```python
rmat
```

```python
signif = pd.DataFrame([rmat['Age Adjusted Rate'],nmat['count']]).transpose()
signif
```

```python
# Find critical values
def crit (r, n):
    return r * np.sqrt(n - 2) / np.sqrt(1 - np.power(r,2))
signif['t_value'] = crit(signif['Age Adjusted Rate'], signif['count'])
signif
```

```python
def pval(t,n):
    return 1.0 - st.t.cdf(np.abs(t),n-1)
signif['p_value'] = pval(signif['t_value'],signif['count'])
signif
```

```python
# On to PLS
# Clean up for PLS
cleandata = data.drop(columns = ['Unnamed: 0','county','code','Deaths','Population','county_and_state_x','state','Crude Rate','county_and_state_y'])
```

```python
cleandata = cleandata.dropna()
cleandata.count()
```

```python
xmat = cleandata.drop(['Age Adjusted Rate'], axis = 1).values
ymat = cleandata['Age Adjusted Rate'].values
```

```python
# Define object and fit
r2 = []
mse = []
for n in range(2,15):
    pls = PLSRegression(n_components=n)
    pls.fit(xmat,ymat)
    # Prediction
    Y_pred = pls.predict(xmat)
    # Calculate scores
    score = r2_score(ymat, Y_pred)
    mserror = mean_squared_error(ymat, Y_pred)
    r2.append(score)
    mse.append(mserror)
```

```python
mp.plot(np.arange(2,15),mse)
```

```python
mp.plot(np.arange(2,15),r2)
```

```python
pls = PLSRegression(n_components=2)
pls.fit(xmat,ymat)
# Prediction
Y_pred = pls.predict(xmat)
# Calculate scores
score = r2_score(ymat, Y_pred)
mserror = mean_squared_error(ymat, Y_pred)
```

```python
pred_df=pd.DataFrame({'fips':xmat[:,0],'predicted_rate':Y_pred[:,0]})
pred_df.head()
```

```python
data2 = data.merge(pred_df,on='fips',how='left')
data2.head()
```

```python
data2['pred_error'] = data2['predicted_rate'] - data2['Age Adjusted Rate']
data2.head()
```

```python
data2['rel_error'] = data2['pred_error'] / data2['Age Adjusted Rate']
data2.head()
```

```python
data2.rel_error.hist(bins=30)
```
