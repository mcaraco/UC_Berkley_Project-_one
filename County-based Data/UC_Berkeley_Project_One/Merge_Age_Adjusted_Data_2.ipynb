```python
infile = 'Source_Data/merged_counties_xnote.csv'
```

```python
data = pd.read_csv(infile)
data.tail()
```

```python
data = data.drop(columns=['Unnamed: 0','Unnamed: 0.1'])
```

```python
# Get rid of notes at end
newdata = newdata.iloc[0:3078,:]
```

```python
data['county_and_state_y'].count()
```

```python
mergedata = data.merge(newdata[['County','Age Adjusted Rate']],left_on='county_and_state_y',right_on='County',how='left')
```

```python
mergedata['Age Adjusted Rate'].count()
```

```python
len(mergedata)
```

```python
mergedata = mergedata.drop(columns='County')
mergedata.tail()
```

```python
#Get rid of the (unreliable) labels from the data using rstrip
mergedata['Age Adjusted Rate'] = mergedata['Age Adjusted Rate'].astype('str')
mergedata['Age Adjusted Rate'] = mergedata['Age Adjusted Rate'].map(lambda x: x.rstrip('()Unreliab '))
```

```python
mergedata.tail()
```

```python
# Turn rate data back to numeric now that strings are removed
mergedata['Age Adjusted Rate'] = pd.to_numeric(mergedata['Age Adjusted Rate'],errors='coerce')
```

```python
mergedata.tail()
```

```python
mergedata.to_csv('Source_Data/merged_counties_2_year_merged.csv')
```
