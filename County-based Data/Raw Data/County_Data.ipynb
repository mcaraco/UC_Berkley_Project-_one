Variables:  "Crude Rate" is the death rate from heart disease (the 'Y' axis) -- some are marked as unreliable.  
'code', 'code2', and 'fips' are all code numbers taht should all match up, likewise the 'county_and_state' names should match up.  I have left them in for now jsut to make sure no one finds an issue that I missed.  
The 'X" varaibles are:
smoking -- adult smoking % from the county health data
obesity -- adult obese % from the county health data
foodenv -- food environment index, "access to healthy foods" 0 - worst, 10 - best, from county health data
exercise -- % of population with access to places for physical activity, from county health data
noinsurance -- % of people under age 65 without health insurance, from county health data (note:  everyone over 65 has Medicare, so they were not counted)
physicians -- primary care physicians per 100,000 population, from county health survey
pollution -- average daily level of PM2.5 (dangerous particulate matter) in micrograms per cubic meter, >12 is unhealthy
diabetes --% of population that is diabetic, from county health data
sunlight -- average daily solar radiation, kJ/sq. m 
blow, ganj, smack -- state-based prevalence data
